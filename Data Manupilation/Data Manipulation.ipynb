## Filter Row Using Operator 
- and (&)
- or (|)

```python
df.iloc[df[(df.Sex == 'female') & (df.Cabin.isnull())].index]
### Using And
```

```python
##  np.logical
df[np.logical_or(df.Age > 30 ,df.Sex =='male') ].index
## we use np.logical
```

```python
d1 = {
    "Name" : ["Karthik" , "Arun" ,"Abishek" ,"Jothi" ,"Dinesh","Mohan"],
    "Age" : [21,22,20,22,21,21],
    "Location" : ["Villupuram" ,"Vikravandi" ,"Agaram" ,"Salem" ,"Gugai","Salem"]
}

d2 = {
    "Name" : ["Karthik" , "Arun" ,"Abishek" ,"Nibi" ,"Aravinth","Kishore"],
    "Age" : [21,22,20,22,21,21],
    "Location" : ["Villupuram" ,"Vikravandi" ,"Agaram" ,"Salem" ,"Nammakal","Namakkal"]
}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
print(df1);print(); df2
```

```python
## inner joint
pd.merge(df1,df2,on="Name") ## It will merge when value are equal
```

```python
## left join
pd.merge(df1,df2,on="Name",how="left")
```

```python
## right join
pd.merge(df1,df2,on="Name",how='right')
```

```python
## Create a New column using square with apply 
df["New_col"] = df.D.apply(lambda x : x**2)
df
```

```python
## create a manual Function for return less than 0
def test(x):
    if(x<0):
        return x
    else:
        return np.log10(x)
```

```python
df['New_column_log10'] = df.D.apply(test)
df
```
