```python
## Base Package 
import pandas as pd
import numpy as np
## Visualization Lib 
import matplotlib.pyplot as plt
%matplotlib inline  
import seaborn as sns
```

```python
## create a Table 
np.random.seed(20)
df = pd.DataFrame(np.random.rand(1000,1),index=pd.date_range('2021-07-24',periods=1000),columns=["Data"])
df
```

```python
df.plot()
```

```python
df.Data.cumsum()
```

```python
## Cumsum do some operation of our data so the value ranges upto 500
df.Data.cumsum().plot()
```

```python
df
```

```python
df["Cum Data"] = df.Data.cumsum()
```

```python
df["Cube Of Data"] = df.Data.apply(lambda x : x**3)
```

```python
df
```

```python
df.plot(figsize = (10,10)) ## Data is not see becozz the data is almost to zero
```

## Histogram
- In Histogram Your x represent the Range
- Y Axis Represent the Frequency

```python
iris.plot(kind="hist")
plt.show()
## This is Overlapping 
```

```python
## See the overlapping
## based on the bins the bar size reduce 
iris.plot(kind="hist" , stacked = True , bins=20,figsize=(8,8),title="Overlapping")
```

```python
## same with horizontal
iris.plot(kind="hist" , stacked = True , bins=20,figsize=(8,8),title="Overlapping",orientation = 'horizontal')
```

```python
iris.describe() ## now we see sepal_width is between 2 to 4
## Sepal width between 2 to 3 there is 100 data 
```

```python
iris.sepal_width.plot(kind = 'hist',title = "Seoal Width") ## it is between 2 to 4.4
## here 2.0 to 2.25  5 occurance 
## 2.25 to 2.5 7 

## here we see it is  normally distributed(GaussianDistribution) 
```

```python
iris.petal_width.plot(kind = 'hist',title = "Petal Width")
## here we see it is not Normally Distribution 
```

```python
### Need Seperate Histogram
iris.hist(figsize = (10,10) ,color = 'red' , alpha =.9) ## Alpha is jus color thickness
plt.show()
```

```python
iris.hist(figsize = (10,10) ,color = 'red' , alpha =.3) ## Alpha is jus color thickness
plt.show()
```
