```python
import pandas as pd
import sqlalchemy
```

## Connect MySql with Python using sqlalchemy

```python
engine = sqlalchemy.create_engine("mysql+pymysql://root:root@localhost:3306/datascience")
```

## read a table in our Mysql DB

```python
df=pd.read_sql_table("student",engine)
df
```

```python
pd.read_sql_table("student",engine,columns=["stud_id","dept"])
```

```python
## Read using Querry
querry="""
select * from student where dept='B.Com';
"""

pd.read_sql_query(querry,engine)
```

## Convert csv File into SQL


```python
carbon=pd.read_csv("carbon_nanotubes.csv",sep=";")
carbon.head()
```

```python
carbon.to_sql(name="Carbon",con=engine,index=False)
```

## read_sql

```python
pd.read_sql("carbon",engine)
```
