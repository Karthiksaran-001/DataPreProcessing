```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',sep='\t')
df.sample(5)
```

```python
## check the DataTypes
df.dtypes
```

```python
df['item_price'].replace(r'\$',' ',inplace=True,regex=True)
## To Remove Dollar in Item Price 
```

```python
df.head()
## Dollar is Removed 
```

```python
df.dtypes
```

```python
change_dtype={
    "item_price":float}
df=df.astype(change_dtype)
## Change the Item_Price Datatypes Object into Float 
```

```python
df.dtypes
## Now item_price Datatypes changes str into float
```

```python
df['order']=str(df['order_id'] * df['quantity'])
## This is Another Method to Change the DataType 
```

```python
df.dtypes
```

```python
## To Drop the Order column
df.drop(['order'],axis=1,inplace=True)
```

```python
df.dtypes
## Now Order Column is Drop 
```

```python
## Now Create a Price based on quantity 
df['Price'] = df['quantity'] * df['item_price']
```

## Pandas GroupBy

```python
stud = pd.read_csv('Mysqldata.csv')
stud.head()
```

```python
group=stud.groupby('dept')
```

```python
## So Now It GroupBy Using Deptartment Column 
for dept,dept_df in group:
    print(dept)
    print(dept_df)
```

```python
## If U Want Specific Group Of DataFrame
group.get_group('B.Com')
```
