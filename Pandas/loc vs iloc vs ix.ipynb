# loc vs iloc vs ix

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df.sample(5)
```

## loc
- Loc helps to filter rows and selecting columns using Labels
- Label means for row "indexes" and for column "columnNames"
## iloc 
- iloc filter remains same but it gives by index wise

```python
## i want 1 row from all the column
df.loc[0,:]
```

```python
df.iloc[0,:]
```

```python
df.loc[[0,1,2],:]
```

```python
df.loc[0:4,:]
```

```python
df.iloc[0:4,:]
```

#### Now we see loc Gives upto 4 but iloc gives upto 3 like stringSlicing

```python
df.loc[:,['Name']]
```

```python
df.iloc[:,['Name']]
## It Gives Error becozz iloc only accept numbers
```

```python
df.iloc[:,[3]]
```

```python
## Gives only Male record from DataFrame
df.loc[df.Sex=='male',:]
```

```python
df.iloc[df[df.Sex=='male'].index]
```
