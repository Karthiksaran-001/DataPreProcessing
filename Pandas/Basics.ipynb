## Read Various option

By default in file first row as consider as an header , if we dont want to consider header = None

```python
pd.read_csv("Mysqldata.csv",header=None)
## Now we see it doesn't consider first row as header
```

We can give Column name as user defined using names we pass a column name using list 

```python
dfs=pd.read_csv("Mysqldata.csv",header=None,names=["A","B","C","D"])
dfs.head()
## Now we change the column name as our own
```

```python
dfs.columns=["One","Two","Three","Four"]
## this is another method to change the column name
```

```python
dfs.head()
```

If we want to skip the rows we use skiprows

```python
pd.read_csv("Mysqldata.csv",skiprows=[0,4,3])
## Now we see it skips the particular row it skip 0 row so 1st row changes into header 
```

Delimeter bydefault csv file seprate by comma if we want to change this we use sep=

```python
carbon=pd.read_csv("carbon_nanotubes.csv",sep=";")
carbon
```
